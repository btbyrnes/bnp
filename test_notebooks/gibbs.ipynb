```python
import numpy as np
from scipy.stats import distributions
from classes.gibbs import gibbs_sample_normal_likelihood
from classes.variables import Normal, Gamma
import matplotlib.pyplot as plt
```

```python
y = distributions.norm.rvs(size=100)
np.mean(y), np.std(y)
```

```python
mu, tau = Normal(0, 3), Gamma()
thetas = [mu, tau]
```

```python
t_ = []
s_ = []
lag = 5
for i in range(2000):
    m, t = gibbs_sample_normal_likelihood(y, [mu, tau])
    if i % lag == 0:
        t_.append(m)
        s_.append(t)

fig, ax = plt.subplots(1, 2, figsize=(12,8))
ax[0].plot(t_)
ax[0].hlines(np.mean(t_), xmin=0, xmax=len(t_), color="red")
ax[1].plot(s_)
ax[1].hlines(np.mean(s_), xmin=0, xmax=len(s_), color="red")

```
