```python
import pymc as pm
import numpy as np
```

```python
y = np.array([-0.12089986,  0.61515674,  0.56811393,  1.07073126,  1.23571253,
      -0.25292445,  0.87784138,  1.78100447,  0.3133824 ,  0.92270987])
```

```python
with pm.Model() as m:
    mu = pm.Normal("mu", mu=0, sigma=1)
    sigma = pm.InverseGamma("sigma", alpha=1, beta=2)

    likelihood = pm.Normal("likelihood", mu=mu, sigma=sigma, observed=y)
```

```python
with m:
    trace = pm.sample()
```

```python
pm.plot_trace(trace)
```

```python
np.mean(trace.posterior["mu"])
```

```python
np.mean(trace.posterior["sigma"])
```
