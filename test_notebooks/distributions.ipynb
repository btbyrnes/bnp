```python
from scipy.stats import distributions as dist
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("bmh")
```

```python
x = np.arange(0, 10, 0.01)
y = dist.gamma.pdf(x, a=2, scale=1)
plt.plot(x,y)
plt.vlines(x=1, ymin=np.min(y), ymax=np.max(y), linestyle="dashed")
plt.title("Gamma ~G(2, 1)")
```

```python
a, s = 1,1
x = np.arange(0, 10, 0.01)
y = dist.gamma.pdf(x, a=a, scale=s)
plt.plot(x,y)
plt.vlines(x=1, ymin=np.min(y), ymax=np.max(y), linestyle="dashed")
plt.title(f"Gamma ~G({a}, {s})")
```

```python
x = np.arange(0, 10, 0.01)
y = dist.invgamma.pdf(x, a=1, scale=2)
plt.plot(x,y)
plt.vlines(x=1, ymin=np.min(y), ymax=np.max(y), linestyle="dashed")
plt.title("Inverse Gamma ~IG(1, 2)")
```

```python
a = 1
scale=1
x = np.arange(0, 10, 0.01)
y = dist.invgamma.pdf(x, a=a, scale=scale)
plt.plot(x,y)
plt.vlines(x=1, ymin=np.min(y), ymax=np.max(y), linestyle="dashed")
plt.title(f"Inverse Gamma ~IG({a}, {scale})")
```
